```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

#### 1.Import file "movierating.csv"

```python
rating=pd.read_csv("movierating1.csv")
rating.head(10)
```

#### 2.	Use “cosine-similarity” to check similarity between users

```python
#create a user list
user_list=[]
for row_index, row in rating.iterrows():
    user=np.array(rating.loc[row_index,rating.columns!="User"]) #remove user index
    user=user.reshape(1,-1)
    user_list.append(user)
user_list[0] #check user 1
```

```python
#compute cosine similatities between all pairs of users
cos_dict={}
for i in range(len(user_list)):
    row=i+1
    for j in range(row,len(user_list)):
        user_pair="user"+str(i+1)+ " & "+"user"+str(j+1)
        cos=cosine_similarity(user_list[i],user_list[j])[0][0]
        cos_dict[user_pair] = cos
```

```python
pd.DataFrame(cos_dict.items(),columns=["user pair","cosine similarity"])
```

#### 3. Find pair of users with the largest and the second largest cosine similarities

```python
largest=max(cos_dict,key=cos_dict.get)
print(f"User pair {largest} has the largest cosine similarity {cos_dict[largest]}")
```

```python
second_largest=list(sorted(cos_dict.values()))[-2]
second_largest_pair=list(cos_dict.keys())[list(cos_dict.values()).index(second_largest)]
print(f"User pair {second_largest_pair} has the second largest cosine similarity {cos_dict[second_largest_pair]}")
```

#### 4.Show the cosine-similarity of all users with user 1 and say who is the most similar to user 1

```python
cos_list=[]
for i in range(len(user_list)):
    cos=cosine_similarity(user_list[0],user_list[i])[0][0]
    cos_list.append(("user1 &"+" user"+str(i+1),cos))

cos_list.sort(key=lambda x:x[1], reverse=True) #sort in descending order 
cos_df=pd.DataFrame(cos_list,columns=["user","cosine_similarity"])
cos_df.head(10)
```
